# subflot
# 여러개의 그래프가 하나에 들어있다고 생각하면 된다.

```python
fig = plt.figure(figsize= (16,10),dpi = 80) # 기본적인 틀을 잡는다.
grid = plt.GridSpec(4,4, hspace = 0.5, wspace = 0.2) # GridSpec는 메인과 옆에 사이즈를 띄워준다. 

ax_main = fig.add_subplot(grid[:-1,:-1]) 
ax_right = fig.add_subplot(grid[:-1, -1], xticklabels = [] , yticklabels = []) #  xticklabels = [] , yticklabels = [] 양 옆의 라벨을 싹다 지운다. 
ax_bottom = fig.add_subplot(grid[-1,0:-1], xticklabels =[], yticklabels= [])

ax_main.scatter('NO2', 'PM10', color = 'r', data =df)
ax_right.hist(df.PM10, 40, orientation = 'horizontal', color = 'lightblue')
ax_bottom.hist(df.NO2, 40, orientation= 'vertical', color = 'lightpink')
```
