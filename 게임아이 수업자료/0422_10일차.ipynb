```python
#CHAMPIONNAME과 TEAMPOSITION을 그룹으로 삼아서 승률을 구하기
```

```python
#그룹으로 묶은뒤에 count() -> 해당 그룹에 있는 모든 데이터의 양
# 그룹으로 묶은뒤에 sum() -> 해당 그룹에 있는 모든 숫자데이터의 합
```

```python
# 이겼으면 1 졌으면 0
```

- win컬럼에 있는 STR형태의 문자열을 숫자로 변경
- 그룹으로 묶는 함수를 사용한 뒤 count함수를 사용해서 WIN컬럼의 데이터 양 구하기
- 데이터의 양이 들어가있는 테이블 하나 생성

- 그룹으로 묶는 함수를 사용한뒤 sum함수를 사용해서 win컬럼의 값을 더한걸 구하기
- 해당 값이 들어가있는 테이블 하나 생성

- cnt_df => 10 sum_df => 5 -> merge cnt/win -> win_per(승률 구하기)

```python
my_df2['WIN'] = my_df2['WIN'].apply(lambda x:1 if x =='True' else 0)
```

```python
cnt_df = my_df2.groupby(['CHAMPIONNAME','TEAMPOSITION'])[['WIN']].count().rename(columns = {"WIN":'CNT'})
```

```python
win_df = my_df2.groupby(['CHAMPIONNAME','TEAMPOSITION'])[['WIN']].sum()
```

```python
cnt_win_df = pd.merge(cnt_df,win_df,on=["CHAMPIONNAME","TEAMPOSITION"])
```

```python
cnt_win_df['win_per'] = round(cnt_win_df.WIN/cnt_win_df.CNT,2)
```

```python
cnt_win_df[cnt_win_df.CNT >5].sort_values(by=['win_per'],ascending = False)[:20]
```
