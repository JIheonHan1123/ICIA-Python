```python
import requests
import pandas as pd
```

```python
url = 'http://openapi.seoul.go.kr:8088/78654862626d6f7436394273514669/json/tbLnOpendataRentV/1/5'
```

```python
url
```

```python
res = requests.get(url).json()
```

```python
row = res['tbLnOpendataRentV']['row']
```

```python
row
```

```python
df = pd.DataFrame(row)
df
```

```python
sel_c = ['SGG_NM','BJDONG_NM','LAND_GBN_NM','FLR_NO','CNTRCT_DE','RENT_GBN','RENT_AREA','RENT_GTN','RENT_FEE','BLDG_NM','BUILD_YEAR']
```

```python
my_df = df[sel_c]
```

```python
my_df.columns=['자치구명','법정동명','지번구분명','층수','계약일','전월세','면적','보증금','임대료','건물명','건축년도']
```

```python
my_df
```

- 전월세 = 전세/월세 따로 뽑아서 2개의 데이터 프레임 만들기
- astype -> 보증금을 astype으로 int, float으로 변경 -> 전세
- astype -> 임대료, 보증금 astype으로 int, float으로 변경 -> 월세
- 분리된 데이터프레임으로 자치구명을 기준으로 groupby를 해서
- 전세는 보증금의 평균/ 월세는 보증금.임대료의 평균을 구하세요
- 보증금 or임대료가 높은 순으로 sort 정렬

```python
rent_df = my_df[my_df.전월세 == '전세']
```

```python
rent_df = rent_df.astype({'보증금':'int'})
```

```python
rent_df = rent_df.astype({'임대료':'int'})
```

```python
type(rent_df['보증금'].iloc[0])
```

```python
group_df = rent_df.groupby(['자치구명'])['보증금'].mean().round(2)
```

```python
group_df.sort_values(by=['보증금'], ascending = False)
```

```python
month_df = my_df[my_df.전월세 == '월세']
```

```python
month_df = month_df.astype({'보증금':'float', '임대료':'float'})
```

```python
group_df2  = month_df.groupby('자치구명')[['보증금', '임대료']].mean().round()
```

```python
group_df2.sort_values(by=['임대료', '보증금'], ascending = False)
```
